```python
cols = ['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
```

```python
cols
```

```python
df = df[cols]
```

```python
df
```

```python
df.columns = ['User_Id' if col == 'Id' else col for col in df.columns]
```

```python
df.columns = ['Date' if col == 'ActivityDate' else col for col in df.columns]
```

```python
df
```

```python
df = df.assign(TotlaHours=round(df.TotalSteps/60, 2))
```

```python
df
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.info()
```

```python
df
```

```python
df['DayOfWeeks'] = df['Date'].dt.day_name()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe(include= 'all')
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['DayOfWeeks'], bins=7, color='blue', width=0.6, rwidth=0.9)
plt.grid(True)

plt.xlabel('Day Of Week')
plt.ylabel('Frequency')
plt.title('Distribution of Day Of Week')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df.TotalSteps , df.Calories, c = df.Calories)

plt.axhline(y=df.Calories.mean(), color='b', linestyle='--')
plt.axvline(x=df.TotalSteps.mean(), color='r', linestyle='--')

plt.xlabel('TotalSteps')
plt.ylabel('Calories')
plt.title('TotalSteps vs Calories')
plt.show()
```

```python
df.describe()
```

```python
plt.figure(figsize = (10,5))
plt.scatter(df.TotlaHours , df.Calories , c = df.Calories)

median_hours = df.TotlaHours.median()
median_calories = df.Calories.median()

plt.axhline(median_calories , color = 'blue' , label = 'median of hours')
plt.axvline(median_hours , color = 'red' , label = 'median of calories')

plt.xlabel('hours')
plt.ylabel('Calories')
plt.title('Calories by total Hours')
plt.legend()
plt.show()
```

```python
VeryActiveMinutes = df.VeryActiveMinutes.sum()
FairlyActiveMinutes = df.FairlyActiveMinutes.sum()
LightlyActiveMinutes = df.LightlyActiveMinutes.sum()
SedentaryMinutes = df.SedentaryMinutes.sum()
```

```python
minutes = [VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes]
labels = ['Very Active Minutes', 'Fairly Active Minutes', 'Lightly Active Minutes', 'Sedentary Minutes']
plt.pie(minutes, labels=labels, autopct='%1.1f%%', explode=(0, 0, 0, 0.1), shadow=True)
plt.title('Minutes of Daily Activity')
plt.show()
```
